### Clean data

#### Missing data in columns

How much missing data is present in each column? There are a few columns that are outliers in terms of the proportion of values that are missing. I will need to identify and document these columns. While some of these columns might have justifications for keeping or re-encoding the data. Besides, I will plot a histogram graph to better visualize the columns with missing data.

```python
def assess_missing_data(df):
    '''
    A function to find out the missing data and its statistic information
    
    input:
    df: the data frame
    
    output:
    df_stats: the statistic information of missing column, will be missing value count and
    missing value percentage
    '''
    
    rec = []
    for column_name in df.columns:
        miss_count = df[column_name].isnull().sum(axis=0)
        miss_percent=miss_count/df.shape[0]
        rec.append([column_name,miss_count,miss_percent*100])
        df_stats = pd.DataFrame.from_records(rec, columns = ['column name', 
                                                             'missing_count','missing_percentage'])  
        df_stats = df_stats.sort_values('missing_percentage',axis=0, ascending = False )
    return df_stats

df_stats = assess_missing_data(df)

df_stats
```

```python
df_stats.plot.bar(x='column name', y='missing_percentage', figsize=(12,6), grid=True, color = '#1f77b4')
plt.title('columns with missing data')
plt.ylabel('percentage of missing data')
plt.xlabel('columns')
```

Columns driver_score, rider_score, arrived_at, cost_distance, cost_duratioon have more than 20% of missing data. However, since this is a very small data set, and we don´t need to train the data with machine learning models, we can keep these columns with a lot of missing data.

```python
def clean_dataframe(df):
    '''
    create a clean dataframe function to use later analysis
    
    input: 
    df: original dataframe
    
    output:
    df_final: dataframe after cleaning it
    '''
    
    df_final = df[(df['price'] != 0) & (df['end_state'] != 'failure')]
    df_final = df_final[(df_final.isnull().sum(axis=1) < 10)]
    
    return df_final

df_clean = clean_dataframe(df)
df_clean.shape
```

### 6. Who used our service most frequent? Are they really our loyal/regular users?

It might be not fair if we only see the average fequency, because there might be users use many times our service per year, there are also might be users only have used once our service. So let´s focus on the users who are very close to our service.

```python
top_users = df_clean.groupby('user_id')['journey_id'].size().reset_index(name='usage_times').sort_values('usage_times',
                                                                                                         ascending = False)
top_users.head(10)
```

```python
# Create the period column to see the time delta between the earliest time and latest time of all 
# the users. 
df_clean['period'] = df_clean.groupby('user_id')['start_at'].transform(lambda x: x.max() -  x.min())
df_period = df_clean[['user_id', 'period']]
df_period = df_period.drop_duplicates('user_id')

# Merge the period with top_users dataframe from before
top_users = top_users.merge(df_period, how = 'left', on = ['user_id'])
top_users['days'] = top_users['period'].dt.days
top_users['intensity'] = top_users['usage_times']/top_users['days']

top_users.head(10)
```

```python
top_users.head(10)
```

```python
top_users[(top_users['days'] != 0)].sort_values('intensity', ascending = False).head(10)
```

You can see at the charts I have showed, not only the frequency that the user use our service but also the intensity they use. When we see the users who have a high frequency of usage, but they have a low intensity along the year, we assume these users are quite loyal to our service, they are regular users during the whole year. But if a users with a very high frequency, like the user 'ae06c34243f741e203e97c48eb42017b', but if you look at the period he used our service, is only 3 months, is very intense. ### 10. Where is the most popular start point? Which area is it?

In this part of analysis, I will explain how to plot the map I want by using geopandas and geographic shapefile. Find out which area or district have most start point. 

```python
# Create a geometry based on the latitude and longitude of the start point.
geometry = [Point(xy) for xy in zip(df['start_lon'], df['start_lat'])]
df['geometry'] = geometry
crs = {'init': 'epsg:4326'}
df_locs = gpd.GeoDataFrame(df, crs=crs, geometry=geometry)
df_locs.head()
```

### 11. Did users pay fair price?

Before analyzing this point, let's first understand the price rules here.

***total price = starting fee + distance price + duration price***

Price distance is the price per kilometer.
price duration is price per minute.

The unit price is different. Usually the duration price is much lower than the distance price. In general, the drive mobile app always take the reference from **Google maps** – assuming the company uses it for route recommendation – **optimizes for time**. Although the optimized routine will take the quickest way, but it might involve more distance in it. Finally the customer might pay even more than expected.

Let's see one example, assum that the duration price is 0.25 euro per minute, and distance price is 1.4 euro per km:

*(6.9 km x 1.40) + (18 minutes x .25) = 14.16 euros.
<br>
(11.0 km x 1.40) + (13 minutes x .25) = 18.65 euros.*

Another trick rule in the pricing is, the price will start to calculate at the time when driver start to drive. It means, if the driver is 1 km, 3 minutes far away from you, you will need also pay for these money as well.

So do the users really **paid a fair price**? I think **NOT**. 
I conclude couple of assumptions to prove my guess.

### Assumption 1: The duration decrease but distance increase
As I said, the time and distance can be really a issue, if the duration decrease and the distance decrease as well, then it is fair, but image if your duration decrease but the distance increase, finally you will pay much more than you should.

The easiest way to do it is find out the correlation between duration and distance

```python
# Firstly let me have a look on the price related data
df_clean[['price','price_distance', 'price_duration', 'distance', 'duration','driver_score']].head()
```

```python
# I doubt the distance here is with a unit of m and duration is with unit of second. Let me first convert them into
# a better format.
df_clean['distance_km'] = df_clean['distance']/1000
df_clean['duration_min'] = df_clean['duration']/60

df_clean[['price','price_distance', 'price_duration', 'distance_km', 'duration_min','driver_score']].head()
```

```python
ax = sns.regplot(x="duration_min", y="distance_km", data=df_clean).set_title('Correlation between duration and distance')
```

Few outliers make this graph more difficult to see. Let's zoom it.

```python
plt.ylim(0,150)
ax = sns.regplot(x="duration_min", y="distance_km", data=df_clean).set_title('Correlation between duration and distance zoom')
plt.ylim(0,150)
```

```python
plt.ylim(0,50)
ax = sns.regplot(x="duration_min", y="distance_km", data=df_clean).set_title('Correlation between duration and distance zoom')
plt.ylim(0,50)
```

Is not difficult to see that the correlation did not have a obvious positive tendency. And there are a very intense cluster in the left area, where the duration is close to 0 but the distance km is climbing up. Which means, when the duration decrease, the distance does not decrease as well. So my 1st assumption is right. When users have a trip with shorter duration but actually **more kilometers are involved, finally you will pay more.**

### Assumption 2: The users paid a lot of money for the kilometers from drivers when they went to pick riders

It is reasonable that the users pay for the distance when driver came to pick them, but when they should pay a lot of distance for it, it is not fair anymore.

From these finding, my guess is approved, the users are paying an unfair price.

Reason 1: The time optimization routine will make the users pay more because the duration decrease but the distance increase.
Reason 2: Users are paying way to much for the distance when driver came to pick them.

The consequence of users are paying unfair money will bring good revenue to the company, but it will let users lose trust on this company, they will think the company is playing dirty game, it will definitely hurt the brand image. ### Recommendations:

**1. Assign more cars or contract more drivers during September to December. And gave more priority to 'asap' start type users.**

There are 23% of trip was not executed. Within this 23%, 75% is because the users cancelled the trip or the trip is failed. *Related business questions: 5, 6*
